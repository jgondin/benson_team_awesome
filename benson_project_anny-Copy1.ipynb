```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python

list_data =[]
list_weeks = ['160102', '160109', '160116','160130','160206','160213',
              '160220','160227','160305','160312','160319','160326','160402']
for u in list_weeks:
    url = 'http://web.mta.info/developers/data/nyct/turnstile/turnstile_' + u + '.txt'
    print(url)
    list_data.append(pd.read_csv(url))

data = pd.concat(list_data)
del list_data
```

```python
#data = pd.read_csv("http://web.mta.info/developers/data/nyct/turnstile/turnstile_160402.txt")
```

```python
data.columns = [column.strip() for column in data.columns]
data.columns
```

```python
data.rename(columns={"EXITS": "EXIT", "ENTRIES": "ENTRY"}, inplace=True)
```

```python
data['dENTRY'] = 0
data['dEXIT'] = 0
```

```python
data['dENTRY'] = data.ENTRY.shift(-1) - data.ENTRY
data['dEXIT'] = data.EXIT.shift(-1) - data.EXIT
data['dTOTAL'] = data.dENTRY + data.dEXIT
```

```python
#diff = data['ENTRY'].iloc[1:].values - data['ENTRY'].iloc[:-1].values
#data['DENTRY'][1:] = diff
#diff = data['EXIT'].iloc[:-1].values - data['EXIT'].iloc[1:].values
#data['DEXIT'][1:] = diff
```

```python
data['DATETIME'] = pd.to_datetime(data.DATE + data.TIME, format='%m/%d/%Y%H:%M:%S')
data['HOUR'] = data['DATETIME'].apply(lambda x: x.hour)
```

```python
#data['WEEKDAY']=data.DATETIME.apply(lambda x: 'Weekday' if 0<=x.weekday()<5 else 'Weekend')
```

```python
# daysofweek = ['Monday','Tuesday','Wednesday','Thursday','Friday','Saturday','Sunday']

# whatdayofweek = lambda x: daysofweek[x.weekday()]


# data['DAYOFWEEK'] = data.DATETIME.apply(whatdayofweek)
```

```python
data_cleaned = data[(data['dENTRY'] >= 0) & (data['dENTRY'] <= 10000) & (data['dEXIT'] >= 0) & (data['dEXIT'] <= 10000)]
```

```python
data_0to4 = data_cleaned[(data_cleaned.HOUR ==22)|(data_cleaned.HOUR ==23)|(data_cleaned.HOUR ==0)|(data_cleaned.HOUR==1)]
```

```python
data_0to4
```

```python
data_grouped = data_0to4.groupby(['DATE','STATION']).sum()
```

```python
data_grouped
```

```python
data_reset = (data_grouped
    .reset_index()
    .sort_values(['dTOTAL'], ascending=(False))
)
data_reset
```

```python
data_reset.columns
```

```python
data_reset['DATE'] = pd.to_datetime(data_reset.DATE, format='%m/%d/%Y')
data_reset['WEEKDAY']=data_reset.DATE.apply(lambda x: 'Weekday' if 0<=x.weekday()<5 else 'Weekend')
```

```python
new_data_grouped = data_reset.groupby(['WEEKDAY','STATION']).sum()
```

```python
new_data_grouped
```

```python
new_data_reset = (new_data_grouped
    .reset_index()
    .sort_values(['dTOTAL'], ascending=(False))
)
new_data_reset
```

```python
# daysofweek = ['Monday','Tuesday','Wednesday','Thursday','Friday','Saturday','Sunday']

# whatdayofweek = lambda x: daysofweek[x.weekday()]


# data_reset['DAYOFWEEK'] = data_reset.DATETIME.apply(whatdayofweek)
```

```python
for r in range(new_data_reset.shape[0]):
    if new_data_reset.ix[r,'WEEKDAY'] =='Weekday':
        new_data_reset.ix[r,'dTOTALACTUAL'] = new_data_reset.ix[r, 'dTOTAL']/(5*13)
    else:
        new_data_reset.ix[r,'dTOTALACTUAL'] = new_data_reset.ix[r, 'dTOTAL']/(2*13)
```

```python
new_data_reset = (new_data_reset
    .sort_values(['dTOTALACTUAL'], ascending=(False))
)
new_data_reset
```
